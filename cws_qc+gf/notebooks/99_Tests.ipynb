```python
# Import necessary libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get the file path from the user
file_path = input("Please enter the file path to the data file: ")

# Load the data
df = pd.read_csv(file_path, sep=";")

# Print the data
print(df)
```

```python
# Format as datetime and set as index the column "date"
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
#Print the data
print(df)
```

```python
# Extract the data for the hours corresponding to August
df_august = df[df.index.month == 8]
# Print the data
print(df_august)
```

```python
print(df_august['70.ee.50.00.0c.e6'])
```

```python
print(df_august['70.ee.50.00.0c.e6'][df_august.index == '2023-08-16 13:00:00'])
```

```python
# Plot the data using matplotlib
# Define size of the plot
plt.figure(figsize=(10,5))
plt.plot(df.index, df)
```
